```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import dataframe_image as dfi
import nest_asyncio
```

```python
A = [42.5, 41.6, 42.1, 41.9, 42.1, 42.2]
B = [39.8, 43.6, 42.1, 40.1, 43.9, 41.9]
C = [43.5, 42.8, 43.8, 43.1, 42.7, 43.3]
D = [35.0, 43.0, 37.1, 40.5, 36.8, 42.2]
E = [42.2, 41.6, 42.0, 41.8, 42.6, 39.0]

geral = [A, B, C, D, E]

mediaReal = 42.0
```

```python
lista_medias = []
analista = ['A', 'B', 'C', 'D', 'E']
lista_dp = []
lista_erro_s = []
coeficiente_de_Variacao = []
exato = []
preciso = []

for lista in geral:

    media = round(np.mean(lista), 2)
    dp = round(np.std(lista, ddof = 1), 2)
    erro_sistematico = round((media - mediaReal)/mediaReal * 100, 2)

    lista_medias.append(media)
    lista_dp.append(dp)
    lista_erro_s.append(erro_sistematico)

    #Analises de precisão e exatidão
    Cv = round(dp/media * 100, 2)
    coeficiente_de_Variacao.append(Cv)

    if Cv > 5 and abs(erro_sistematico) > 5:
        exato1 = 'Não exato'
        precisao1 = 'Não preciso'
        exato.append(exato1)
        preciso.append(precisao1)
    elif Cv > 5 and abs(erro_sistematico) < 5:
        exato2 = 'Exato'
        precisao2 = 'Não preciso'
        exato.append(exato2)
        preciso.append(precisao2)
    elif Cv < 5 and abs(erro_sistematico) > 5:
        exato3 = 'Não exato'
        precisao3 = 'Preciso'
        exato.append(exato3)
        preciso.append(precisao3)
    elif Cv < 5 and abs(erro_sistematico) < 5:
        exato4 = 'Exato'
        precisao4 = 'Preciso'
        exato.append(exato4)
        preciso.append(precisao4)
    
dados = {
    'Analista' : analista,
    'Média' : lista_medias,
    'Desvio padrão' : lista_dp,
    'Erro sistemático (%)' : lista_erro_s,
    'Coeficiente de variação (%)' : coeficiente_de_Variacao,
    'Exato ?': exato,
    'Preciso ?': preciso
}
```

```python
df = pd.DataFrame(dados)
df
```
